```python
import numpy as np
import pandas as pd
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from glob import glob
```

```python
femalepath = glob('./data/female/*.jpg')
malepath = glob('./data/male/*.jpg')
```

```python
len(femalepath),len(malepath)
```

```python
img = cv2.imread(femalepath[0])
```

```python
plt.imshow(img)
```

```python
gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
gray.shape
```

```python
haar = cv2.CascadeClassifier('./model/haarcascade_frontalface_default.xml')
faces = haar.detectMultiScale(gray,1.5,5)
faces
```

```python
for x,y,w,h in faces:
    crop_img = img[y:y+h,x:x+w]
plt.imshow(crop_img)
cv2.imwrite('./data/crop/male/{}_{}.jpg'.format("f",0),crop_img)
```

```python
def extract_crop_images(path,gender,i):
    img = cv2.imread(path)
    gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
    faces = haar.detectMultiScale(gray,1.5,5)
    for x,y,w,h in faces:
        crop = img[y:y+h,x:x+w]
        if gender=='male':
            cv2.imwrite('./data/crop/male/{}_{}.jpg'.format(gender,i),crop)
        else:
            cv2.imwrite('./data/crop/female/{}_{}.jpg'.format(gender,i),crop) 
        
```

```python
for i,path in enumerate(femalepath):
    try:
        extract_crop_images(path,'female',i)
        print('INFO: {}/{} cropped sucessfully'.format(i,len(femalepath)))
    except:
        print('INFO: {}/{} not cropped sucessfully'.format(i,len(femalepath)))
```

```python
extract_crop_images(femalepath[0],'female',1)
```

```python
for i,path in enumerate(malepath):
    try:
        extract_crop_images(path,'male',i)
        print('INFO: {}/{} cropped sucessfully'.format(i,len(malepath)))
    except:
        print('INFO: {}/{} not cropped sucessfully'.format(i,len(malepath)))
```
